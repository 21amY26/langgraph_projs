# HELLO WORLD AGENT

```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph

''' 
StatGraph is a framework that helps design and manage flow of application in graph
'''
```

```python
#creating AgentState - shared data struct that keeps track of info as appln. runs

class AgentState(TypedDict):
    message : str

def greeting_node(state: AgentState) -> AgentState:
    ''' 
    simple node that adds a greeting message to the state
    '''

    state['message'] = "Bonjour! " + state['message'] + ".\n Did you know this is my first agent?"

    return state
```

```python
graph = StateGraph(AgentState)

graph.add_node('greeter', greeting_node)

graph.set_entry_point('greeter')
graph.set_finish_point('greeter')

app = graph.compile()
```

```python
from IPython.display import Image,display
display(Image(app.get_graph().draw_mermaid_png()))
```

```python
result = app.invoke({'message' : 'Amy'})
```

```python
result['message']
```

# Personalized Compliment Agent

```python
class AgentStateV2(TypedDict):
    name : str

def compliment_node( state: AgentStateV2) -> AgentStateV2:
    return {'name' : state['name'] + ", you are doing great!"}
```

```python
graph2 = StateGraph(AgentStateV2)
graph2.add_node('compliment', compliment_node)
graph2.set_entry_point('compliment')
graph2.set_finish_point('compliment')

app2 = graph2.compile()
```

```python
display(Image(app2.get_graph().draw_mermaid_png()))
```

```python
result = app2.invoke({'name' : 'Amelia'})
```

```python
result['name']
```
